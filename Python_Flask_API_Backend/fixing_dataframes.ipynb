```python
import pandas as pd
import pickle
```

```python
with open('data.pickle', 'rb') as f:
    # The protocol version used is detected automatically, so we do not
    # have to specify it.
    matchList = pickle.load(f)
```

```python
# df1 = matchDF[0::2]
# df2 = matchDF[1::2]
# df2
for i, match in enumerate(matchList):
    for j, participant in enumerate(match['info']['participants']):
        matchList[i]['info']['participant' + str(j)] = participant
    matchList[i]['info'].pop('participants')
    matchList[i]['metadata'].update(matchList[i]['info'])
    matchList[i].pop('info')
```

```python
# df1[['participant1', 'participant2', 'participant3', 'participant4', 'participant5', 'participant6', 'participant7', 'participant8']]
df = pd.DataFrame()# match DataFrame
for match in matchList:
    matchdf = pd.DataFrame(match)
    matchdf = matchdf.T
    df = df.append(matchdf)
df.info()
```

```python
df1 = matchDF[0::2]
df2 = matchDF[1::2]
matchesDF = pd.combine(df1, df2)
```
